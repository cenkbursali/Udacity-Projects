### Which genres are most popular from year to year?

First we check all genres. Cell in `genres` column have multiple values seperated by pipe ("|"). Therfore, we should split multiple values by pipe.

```python
all_genres = movies['genres'].str.split("|").str[0].unique()
all_genres
```

There are 20 different genres. We sort genres according to their popularity. 

```python
genre_dict = {}

for genre in all_genres:
    genre_dict[genre] = movies["popularity"][movies['genres'].str.contains(genre)].mean()
    
{k: v for k, v in sorted(genre_dict.items(), key=lambda item: item[1], reverse=True)}
```

We create a dataset keeping the most popular three genres in each year.

```python
genre_year = pd.DataFrame()

for genre in all_genres:
    genre_year[genre] = movies[["popularity", "release_year"]
                              ][movies['genres'].str.contains(genre)].groupby("release_year")["popularity"].mean()

genre_year = genre_year.apply(lambda s: s.abs().nlargest(3).index.tolist(), axis=1)

genre_year
```

We reorganize this dataset for better visualization.

```python
popular_genres = pd.concat([genre_year.str[0], genre_year.str[1], genre_year.str[2]], axis=1).rename(columns={0: "First", 
                                                                                             1: "Second", 2:"Thrid"})

popular_genres
```

### What are the most popular genres in 2000s?

In the previous question, we obtained a data set showing the 3 most popular genres in each year. We use this dataset to find the most popular genres of the 2000s now. 

```python
popular_genres.iloc[40:,:].apply(pd.Series.value_counts).sum(axis=1).plot.pie(labeldistance=None, legend=True)
plt.legend(frameon=False, bbox_to_anchor=(1.5,.85))
plt.ylabel("");
plt.title("The most popular genres in 2000s")
```

The most liked genres by the audience in the 2000s are adventure, fantasy and science fiction.

### What kinds of properties are associated with movies that have high revenues?

The original data set contains the `revenue` and `revenue_adj` columns. We kept the `revenue_adj` column instead of `revenue`. Thus, we made a comparison with the present values of their revenues. Firstly, we look descriptive summaries of `revenue` column.

```python
movies["revenue_adj"].describe()
```

It seems that the revenue of the majority of films is 0. This makes descriptive statistics of whole column meaningless. For this reason, let's take a look at the descriptive statistics again by removing the films with 0 revenue.

```python
movies["revenue_adj"][movies["revenue_adj"]!=0].describe()
```

Analysis od `revenue` column is difficult because it contains so many unique values. So let's divide this column into categories. Since most of the films have a revenue of 0, let's group them in a separate group.

```python
# Quantile-based categorization of movies with revenues. 
a = pd.qcut(movies["revenue_adj"][movies["revenue_adj"]!=0], 4, labels=["low", "medium", "high", "very_high"])

# Categorization of movies without revenue
b = movies.query("revenue_adj==0")["revenue_adj"].astype("str")

movies["revenue_cat"] = pd.concat([a,b]).replace("0.0", "zero_revenue")
```

First, we look at the revenue-budget relation.

```python
plot1 = movies.groupby("revenue_cat")[
    "budget_adj"].mean().sort_values(ascending=False).plot.bar(rot=0);

plt.xlabel("Revenue");
plt.ylabel("Budget");
plt.title("Revenue vs. Budget");
```

High-budget movies seem to get higher grossing.

We look at the ratings that the movies get from the users on the basis of revenue.

```python
movies.groupby("revenue_cat")["vote_average"].mean().sort_values(ascending=False).plot.bar(rot=0);

plt.xlabel("Revenue");
plt.ylabel("Average Vote");
plt.title("Revenue vs. Average Vote");
```

It can be said that movies with higher revenues are more liked. However, the revenue-rating relation is weaker than the revenue-budget relationship.

Finally, let's evaluate the popularity of movies based on their revenues.

```python
movies.groupby("revenue_cat")["popularity"].mean().sort_values(ascending=False).plot.bar(rot=0);

plt.xlabel("Revenue");
plt.ylabel("Populartiy");
plt.title("Revenue vs. Popularity");
```

A plot quite similar to the budget-revenue is obtained. The popularity of the movies is directly related to its revenue.
